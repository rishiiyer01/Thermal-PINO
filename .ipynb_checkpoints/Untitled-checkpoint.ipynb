```python
X_data=np.load("/home/iyer.ris/Pipe_X.npy")
Y_data=np.load("/home/iyer.ris/Pipe_Y.npy")

randPipeX=X_data[45]
randPipeY=Y_data[45]
print(randPipeY.shape)
#griddy = tensor(np.stack((randPipeX, randPipeY), -1), 
                   #channel(vector=['x', 'y'])) 
# Initialize the list for storing quadrilaterals
quadrilaterals = []
num_rows, num_cols = randPipeX.shape

# Flatten the mesh grid
points = np.stack((randPipeX.flatten(), randPipeY.flatten()), axis=-1)

# Create quadrilaterals
for i in range(num_rows - 1):
    for j in range(num_cols - 1):
        # Indices of vertices for the current quadrilateral
        v1 = i * num_cols + j
        v2 = v1 + 1
        v3 = v1 + num_cols
        v4 = v3 + 1

        # Append the quadrilateral (v1, v2, v4, v3) to the list
        quadrilaterals.append([v1, v2, v4, v3])

# Call the provided function
polygon_surface = _poly_surface.surface_from_numpy(points, quadrilaterals)
# ... (previous code for polygon_surface and lies_inside_polygon) ...
# ... (previous code for creating quadrilaterals and polygon_surface) ...


from phi.math import spatial

def lies_inside_polygon(point, polygon):
    """Checks if a point lies inside a polygon using the ray casting algorithm.
    
    Args:
        point: (x, y) coordinates of the point.
        polygon: List of (x, y) vertex coordinates in the polygon.
    
    Returns:
        True if the point is inside the polygon, False otherwise.
    """
    x, y = point
    num_intersections = 0
    for i in range(len(polygon)):
        j = (i + 1) % len(polygon)  # Loop around to the first vertex
        p1, p2 = polygon[i], polygon[j]
        if ((p1[1] > y) != (p2[1] > y)) and \
           (x < (p2[0] - p1[0]) * (y - p1[1]) / (p2[1] - p1[1]) + p1[0]):
            num_intersections += 1
    return num_intersections % 2 == 1


nx = 50
ny = 50

# Get the minimum and maximum coordinates of the polygon vertices
min_x, min_y = polygon_vertices.min(axis=0)
max_x, max_y = polygon_vertices.max(axis=0)

# Create a Box object representing the bounds of the grid
bounds = Box['x,y', min_x:max_x, min_y:max_y]

# Create a Shape object for the grid resolution
resolution = spatial(x=nx, y=ny)

# Create a CenteredGrid with the specified bounds and resolution
grid = CenteredGrid(resolution=resolution, extrapolation=extrapolation.ZERO, bounds=bounds)

# Get the grid points as a tensor
grid_points = grid.points.vector['x', 'y']

# Convert the grid_points tensor to a NumPy array with explicit dimension order
grid_points_np = grid_points.numpy(order=('x', 'y', 'vector'))

# Convert the NumPy array to a PyTorch tensor
grid_points_torch = torch.from_numpy(grid_points_np)

# Convert the polygon_surface to a NumPy array with explicit dimension order
polygon_vertices = polygon_surface.vertices.numpy(order=('vertices', 'vector'))

# Create a mask for points inside the polygon
inside_mask = torch.zeros_like(grid_points_torch[:, :, 0], dtype=torch.bool)

for i in range(nx):
    for j in range(ny):
        point = grid_points_np[i, j]
        if lies_inside_polygon(point, polygon_vertices):
            inside_mask[i, j] = True

# Set the values of the grid based on the inside_mask
grid_values = torch.where(inside_mask, 1.0, 0.0)

# Convert the grid_values back to a phi.math.Tensor
grid_values_phi = math.tensor(grid_values, grid.shape)

# Update the grid with the new values
grid = grid.with_values(grid_values_phi)
```
