```python
X_data = np.load("/home/iyer.ris/Pipe_X.npy")
Y_data = np.load("/home/iyer.ris/Pipe_Y.npy")

randPipeX = X_data[45]
randPipeY = Y_data[45]
print(randPipeY.shape)

def lies_inside_polygon(points, polygon_x, polygon_y):
    """Checks if points lie inside a polygon using the ray casting algorithm."""
    x, y = points[:, 0], points[:, 1]
    num_intersections = np.zeros_like(x, dtype=int)
    for i in range(len(polygon_x)):
        j = (i + 1) % len(polygon_x)
        intersect = ((polygon_y[i] > y) != (polygon_y[j] > y)) & \
                    (x < (polygon_x[j] - polygon_x[i]) * (y - polygon_y[i]) / (polygon_y[j] - polygon_y[i]) + polygon_x[i])
        num_intersections[intersect] += 1
    return num_intersections % 2 == 1

nx = 50
ny = 50

# Get the minimum and maximum coordinates of the original meshgrid
min_x, min_y = randPipeX.min(), randPipeY.min()
max_x, max_y = randPipeX.max(), randPipeY.max()

# Create a Box object representing the bounds of the grid
bounds = Box['x,y', min_x:max_x, min_y:max_y]

# Create a Shape object for the grid resolution
resolution = spatial(x=nx, y=ny)

# Create a CenteredGrid with the specified bounds and resolution
grid = CenteredGrid(resolution=resolution, extrapolation=extrapolation.ZERO, bounds=bounds)

# Get the grid points as a tensor
grid_points = grid.points.vector['x,y']

# Convert the grid_points tensor to a NumPy array with explicit dimension order
grid_points_np = grid_points.numpy(order=('x', 'y', 'vector'))

# Create a mask for points inside the polygon
inside_mask = lies_inside_polygon(grid_points_np.reshape(-1, 2), randPipeX.flatten(), randPipeY.flatten())
inside_mask = inside_mask.reshape(nx, ny)

# Set the values of the grid based on the inside_mask
grid_values = math.tensor(inside_mask, grid.shape)

# Update the grid with the new values
grid = grid.with_values(grid_values)
```
